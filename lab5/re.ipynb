```python
from re import *
with open("row.txt", "r", encoding="utf-8") as file:
    info = file.read().splitlines() 
```

```python
print(info)
```

```python
#ex1
patt = r"(a)(b) *"
for item in info:
    if search(patt,item):
        print(item)
```

```python
#ex2
patt = "ab{2,3}"
for item in info:
    if search(patt,item):
        print(item)
```

```python
#ex3
patt = r"[a-z]+_[a-z]+"
for item in info:
    if search(patt,item):
        print(item)
```

```python
#ex4
patt = r"[A-Z][a-z]+"
for item in info:
    if search(patt,item):
        print(item)
```

```python
#ex5
patt = r"a.*b" #жадный поиск
for item in info:
    if match(patt,item):
        print(item)
```

```python
#ex6

info6 = info
patt = r"[ ,\.]"
for item in info6:
    print(sub(patt,":",item))
```

```python
#ex6

item = "alsfdh."
patt = r"[\.]"

print(sub(patt,":",item))
```

```python
#ex7
info7 = info.copy()
for item in range(len(info7)):
    snake = info7[item].split() #разделяю строки на слова 
    for i in range(len(snake)): #проход по словам с помощью индекса
        if search("_",snake[i]):  #проверка snake или нет 
            snakelist = snake[i].split("_") # snake слово разделяю на отдельные слова по _ (превращяю в  лист) 
            snakelist = [word.capitalize() for word in snakelist] #цикл первую букву зашлавной делает слова
            info7[item] = "".join(snakelist)
    
  
print(info7[24])
            
        
```

```python
#ex9
patt = r"([A-Z])"
info9=info
for item in info9:
    if search(patt,item):
        print(sub(patt,r"\1 ",item))
    

        
```

```python
#ex10
info10 = info.copy()
patt = r"([A-Z])"
for item in range(len(info10)): #по индексам
    camel = info10[item].split()  #разделяю строку на слова 
    for i in range(len(camel)):  #длина camel слова 
        if search(patt,camel[i]): #проверка есть ли camel
            camel[i] = sub(patt,r"\1_",camel[i]).lower() #замена заглавных букв на те же буквы но с _ ,  \1 это ссылка на группу
         
    info10[item] = " ".join(camel) # сшиваю лист в слово через пробел
print(info10)
            
        
```
